```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
```

```python
mul_pov = pd.read_excel("multidimensional_poverty.xlsx", skiprows=2)
mul_pov
```

```python
mul_pov.rename(columns={'Unnamed: 0':'Region', 'Unnamed: 1':'Country code',
                        'Unnamed: 2':'Economy', 'Unnamed: 3':'Reporting year',
                        'Unnamed: 4':'Survey name','Unnamed: 5':'Survey year',
                        'Unnamed: 6':'Survey coverage', 'Unnamed: 7':'Welfare type',
                        'Unnamed: 8':'Survey comparability', 'Unnamed: 15':'Multidimensional poverty headcount ratio (%)'},
              inplace=True)
```

```python
mul_pov
```

```python
mul_pov.replace("-", np.nan, inplace=True)
```

```python
mul_pov = mul_pov.dropna()
mul_pov
```

```python
hiv_data = pd.read_csv("HIV data 2000-2023.csv", encoding='ISO-8859-1', thousands=' ')
hiv_data['Value'] = hiv_data['Value'].str.extract(r'(\d[\d\s]+)')[0].str.replace(' ', '').astype(float)
hiv_data
```

```python
hiv_data = hiv_data.dropna()
hiv_data
```

```python
# Make sure Period and Reporting year are numeric
hiv_data['Period'] = pd.to_numeric(hiv_data['Period'])
mul_pov['Reporting year'] = pd.to_numeric(mul_pov['Reporting year'])

# Merge on Country Code and Year
merged_df = pd.merge(
    hiv_data,
    mul_pov,
    how='left',
    left_on=['SpatialDimValueCode', 'Period'],
    right_on=['Country code', 'Reporting year']
)
```

```python
merged_df
```

```python
# Drop rows where 'Multidimensional poverty headcount ratio (%)' is missing
clean_df = merged_df.dropna(subset=["Multidimensional poverty headcount ratio (%)"]).reset_index(drop=True)
```

```python
# Correlation Check
correlation = clean_df['Value'].corr(clean_df['Multidimensional poverty headcount ratio (%)'])
print(f"Correlation between HIV and Poverty headcount: {correlation:.2f}")
```

```python
clean_df = clean_df.rename(columns={
    "Monetary (%)": "monetary",
    "Educational attainment (%)": "educational_attainment",
    "Educational enrollment (%)": "educational_enrollment",
    "Electricity (%)": "electricity",
    "Sanitation (%)": "sanitation",
    "Drinking water (%)": "drinking_water",
    "Multidimensional poverty headcount ratio (%)": "poverty_headcount",
    "Value": "hiv_cases"
})

# Optional: Drop any rows that still have missing values in the predictors
model_df = clean_df.dropna(subset=[
    "poverty_headcount",
    "monetary",
    "educational_attainment",
    "educational_enrollment",
    "electricity",
    "sanitation",
    "drinking_water"
])

# Mixed Effects Model
expanded_model = smf.mixedlm(
    "hiv_cases ~ poverty_headcount + electricity + sanitation + drinking_water + educational_attainment + educational_enrollment + monetary",
    model_df,
    groups=model_df["Location"],
    re_formula="~Period"
)
expanded_result = expanded_model.fit()

# Show results
print(expanded_result.summary())

```
