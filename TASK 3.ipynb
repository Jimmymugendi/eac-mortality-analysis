```python
import geopandas as gpd
import os

# Directory where you saved your country shapefiles
folder = "C:/Users/pc/Desktop/Cema task 1/shapefiles_eac"

# List shapefile paths for each EAC country
shapefiles = [
    os.path.join(folder, 'BDI_Adm0.shp'),  # Burundi
    os.path.join(folder, 'COD_Adm0.shp'),  # Congo DRC
    os.path.join(folder, 'KEN_Adm0.shp'),  # Kenya
    os.path.join(folder, 'RWA_Adm0.shp'),  # Rwanda
    os.path.join(folder, 'SSD_Adm0.shp'),  # South Sudan
    os.path.join(folder, 'TZA_Adm0.shp'),  # Tanzania
    os.path.join(folder, 'UGA_Adm0.shp'),  # Uganda
    os.path.join(folder, 'SOM_Adm0.shp'),  # Somalia
]

# Load and combine them
gdfs = [gpd.read_file(shp) for shp in shapefiles]
eac_gdf = gpd.GeoDataFrame(pd.concat(gdfs, ignore_index=True))

# Plot to check
eac_gdf.plot(figsize=(10, 10), edgecolor='black')
plt.title('EAC Countries Map')
plt.show()
```
